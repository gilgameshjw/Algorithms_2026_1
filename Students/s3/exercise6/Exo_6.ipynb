## 2.	Generate a tree of depth N = 3, with initial parent tree of weight 1/n

```python
class A_Tree:
    def __init__(self, weight):
        self.weight = weight
        self.children = []

def generate_tree(n, current_depth, max_depth, current_weight):
    node = A_Tree(current_weight)

    # to generate children nodes until matches the max depth by recurence
    if current_depth < max_depth:
        child_weight = current_weight / n
        for _ in range(n):
            child_node = generate_tree(n, current_depth + 1, max_depth, child_weight)
            node.children.append(child_node)            
    return node

# suppose a random number for n
n = 5
# give the required max depth 3
max_depth = 3  
current_weight = 1 / n

root = generate_tree(n, 1, max_depth, current_weight)
```

## 3.	create a depth ﬁrst recursive function visiting each node and summing up the weights. Make sure it returns 1 for various n’s

```python
def dfs_sum(node, current_depth, target_depth):
    if current_depth == target_depth:
        return node.weight
    
    total = 0
    for child in node.children:
        #sum up weights for each child node
        total += dfs_sum(child, current_depth + 1, target_depth)
    return total

n = 5
N = 3
# the total original weight shouold be 1 if to mactch the requirment 'it returns 1 for various n’s'
root = generate_tree(n, 1, N, 1)

# call
result = dfs_sum(root, 1, 3)
print(result)
```

## 4.	Same with breadth ﬁrst, check also 1

```python
from collections import deque

def bfs_sum(root):
    total = 0
    queue = deque([root])
    while queue:
        node = queue.popleft()
        total += node.weight
        queue.extend(node.children)
    return total

result = bfs_sum(root)
print(result)
```

```python
# correction--only to sum up the deepest layer
def bfs_sum_fixed(root):
    # 我们只想要最后一层的总和，因为它应该等于初始能量
    queue = deque([(root, 1)]) # 记录节点和深度
    total_at_last_level = 0
    max_d = 3 # 题目要求的深度
    
    while queue:
        node, depth = queue.popleft()
        if depth == max_d:
            total_at_last_level += node.weight
        for child in node.children:
            queue.append((child, depth + 1))
            
    return total_at_last_level
result1 = bfs_sum_fixed(root)
print(result1)
```

## 6.	Write recursive and non recursive versions of breadth ﬁrst.

```python
# Non-recursive (Iterative) BFS
def bfs_iterative(root):
    res = 0
    q = deque([root])
    while q:
        node = q.popleft()
        res += node.weight
        q.extend(node.children)
    return res
```

```python
# Recursive BFS (Level-order recursion)
def bfs_recursive(queue, current_sum=0):
    if not queue:
        return current_sum
    
    node = queue.popleft()
    current_sum += node.weight
    queue.extend(node.children)
    
    return bfs_recursive(queue, current_sum)
```

## 7.	Whilst depth ﬁrst can be implemented recursively, it is not recommended for breadth ﬁrst! Explain why?

### Bcs recursion uses the Call Stack, which is a Last-In-First-Out (LIFO) structure. This perfectly matches the behavior of DFS. For BFS, relies on a First-In-First-Out (FIFO) logic. To force BFS into a recursive structure, will have to pass a manual queue through the recursion. This offers no benefit over a simple while loop and may lead to a stack overflow for very large trees, as every single node visit adds a frame to the stack.
